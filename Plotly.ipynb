```python
tips = sns.load_dataset('tips') #hotel tips
```

```python
tips
```

```python
sns.relplot(x='total_bill',y='tip',data=tips) #realplot
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker') #smoker is a category
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='day',style='smoker')  
#hue = color, style = shape
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,style='size')
```

```python
 sns.relplot(x='total_bill',y='tip',data=tips,hue='size')
```

```python
sns.catplot(x='day',y='total_bill',data=tips) #categorical vs numerical data
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm') 
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```

```python
#who has given a maximum tips
sns.catplot(x='day',y='total_bill',kind='box',hue='sex',data=tips)
```
